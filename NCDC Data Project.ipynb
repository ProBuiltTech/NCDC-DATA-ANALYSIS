```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
from urllib.request import Request,urlopen
#>. this is use to read a file from a website
# n
```

```python
req = Request('https://covid19.ncdc.gov.ng/',
             headers={'user-agent':'mozilla/5.0'})
```

```python
web= urlopen(req).read()
```

```python
ncdc= pd.read_html(web)[0] #>.> this is used to call what is in the first index in the website
```

```python
ncdc
```

```python
ncdc.to_excel('Covid19 statistics In Nigeria.xlsx',sheet_name= 'DATA')
```

```python
ax,fig= plt.subplots(figsize=(8,10),dpi=800)
sns.barplot(x='No. of Cases (Lab Confirmed)',y='States Affected',data=ncdc)
plt.tight_layout()
plt.title('Number of Confirmed Cases In Nigeria')
plt.savefig('Number of Confirmed Cases In Nigeria.png')
plt.show()
```

```python
ax,fig= plt.subplots(figsize=(8,10),dpi=800)
sns.barplot(x='No. of Cases (on admission)',y='States Affected',data=ncdc)
plt.tight_layout()
plt.title('Number of Cases on admission In Nigeria')
plt.savefig('Number of Cases on admission In Nigeria.png')
plt.show()
```

```python
ax,fig= plt.subplots(figsize=(8,10),dpi=800)
sns.barplot(x='No. Discharged',y='States Affected',data=ncdc)
plt.tight_layout()
plt.title('Number of Discharged In Nigeria')
plt.savefig('Number Of Discharged In Nigeria.png')
plt.show()
```

```python
ax,fig= plt.subplots(figsize=(8,10),dpi=800)
sns.barplot(x='No. of Deaths',y='States Affected',data=ncdc)
plt.tight_layout()
plt.title('Number of Deaths In Nigeria')
plt.savefig('Number of Deaths In Nigeria.png')
plt.show()
```

```python
# state with the highest confirmed cases:

highest= np.max(ncdc['No. of Cases (Lab Confirmed)'])
```

```python
print(f'The state with highest Number Cases is LAGOS which has {highest} cases')
```

```python
# state with the lowest confirmed cases:

lowest= np.min(ncdc['No. of Cases (Lab Confirmed)'])
```

```python
print(f'The states with lowest confirmed cases is kogi states\n which has {lowest} confirmed cases')
```

```python
#: No. of Cases (Lab Confirmed) that >=5000:

data= ncdc[(ncdc['No. of Cases (Lab Confirmed)']>=5000)]
```

```python
#: No. of Cases (Lab Confirmed) that >=5000:

ax,fig= plt.subplots(figsize=(8,4),dpi=800)
sns.barplot(x='States Affected',y='No. of Cases (Lab Confirmed)',data=data)
plt.title('No. of Cases (Lab Confirmed) that >=5000')
plt.savefig('No. of Cases (Lab Confirmed)greater than or equal to 5k.png')
plt.show()
```

```python
exe= ncdc[(ncdc['No. of Cases (Lab Confirmed)']>=2000)&
          (ncdc['No. of Cases (Lab Confirmed)']<=5000)]
```

```python
#: No. of Cases (Lab Confirmed) that >=2000 & <=5000:

ax,fig= plt.subplots(figsize=(8,4),dpi=800)
sns.barplot(x='States Affected',y='No. of Cases (Lab Confirmed)',data=exe)
plt.title('No. of Cases (Lab Confirmed) that >=2000 & <=5000')
plt.savefig('No. of Cases (Lab Confirmed) greater than 2k and less than 5k.png')
plt.show()
```
